### Exercici 1 (Nivell 1)
#### Realitza web scraping d'una pàgina de la borsa de Madrid (https://www.bolsamadrid.es) utilitzant BeautifulSoup i Selenium.

```python
#scrapping using Beautiful Soup

#we want to scratch share prices from IBEX35
url = "https://www.bolsamadrid.es/esp/aspx/Mercados/Precios.aspx?indice=ESI100000000"
page = requests.get(url)

soup = BeautifulSoup(page.content, "html.parser")

#share prices are in a table, we use the id selection
results = soup.find(id="ctl00_Contenido_tblAcciones")
```

```python
#from the results we take all td elements (data from table)
table_items = results.find_all("td")                                

#we make a list with elements from each column in the table
#probably this script can be improved and it is not necessary all this code 
#but at the moment, it works and I do not have time to improve it

Nombre_list = []
length = len(table_items)
i = 0
while i < length:
    Nombre_list.append(table_items[i].text.strip())
    i += 9

Ultima_list = []
length = len(table_items)
i = 1
while i < length:
    Ultima_list.append(table_items[i].text.strip())
    i += 9
    
Dif_list = []
length = len(table_items)
i = 2
while i < length:
    Dif_list.append(table_items[i].text.strip())
    i += 9
    
Max_list = []
length = len(table_items)
i = 3
while i < length:
    Max_list.append(table_items[i].text.strip())
    i += 9
    
Min_list = []
length = len(table_items)
i = 4
while i < length:
    Min_list.append(table_items[i].text.strip())
    i += 9
    
Volumen_list = []
length = len(table_items)
i = 5
while i < length:
    Volumen_list.append(table_items[i].text.strip())
    i += 9

Euro_list = []
length = len(table_items)
i = 6
while i < length:
    Euro_list.append(table_items[i].text.strip())
    i += 9    

Fecha_list = []
length = len(table_items)
i = 7
while i < length:
    Fecha_list.append(table_items[i].text.strip())
    i += 9        
    
Hora_list = []
length = len(table_items)
i = 8
while i < length:
    Hora_list.append(table_items[i].text.strip())
    i += 9    

#we make a dataframe with all columns    
columns = ['Nombre','Últ.','% Dif','Máx.','Mín',
           'Volumen','Efectivo','Fecha','Hora']
    
df = pd.DataFrame(list(zip(Nombre_list, Ultima_list, Dif_list, Max_list,
                          Min_list, Volumen_list, Euro_list, 
                          Fecha_list, Hora_list)), columns = columns)

df
```

```python
#now we repeat but scrapping using Selenium

#we use this proxy because we have server rejection issues
PROXY = "11.456.448.110:8080"
firefox_options = Options()
firefox_options.add_argument('--proxy-server=%s' % PROXY)

#headless browser
firefox_options.headless = True
driver = webdriver.Firefox(options=firefox_options)
```

```python
#url with shares prices from IBEX35
url = 'https://www.bolsamadrid.es/esp/aspx/Mercados/Precios.aspx?indice=ESI100000000'
driver.get(url)
```

```python
#identifying elements by id and taking text
table_items = driver.find_element_by_id("ctl00_Contenido_tblAcciones").text
print(table_items)

#closing driver
driver.quit()
```
